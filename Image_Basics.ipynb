```python
import cv2
```

```python
path_3ri = r"/content/3RI logo.png"
path_3 = r"/content/Example_3.jpg"
path_nature = r"/content/Nature.jpg"
```

## By default, images are displayed in BGR format by Computer vision library. Convert them to RGB if you want to look at the image as displayed in pc

```python
img1 = cv2.imread(path_3ri)
img1
```

```python
img1_rgb = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img1_rgb
```

```python
img2 = cv2.imread(path_3)
img2
```

```python
img2_grayscale = cv2.imread(path_3,cv2.IMREAD_GRAYSCALE)
img2_grayscale
```

```python
img_nature = cv2.imread(path_nature)
img_nature
```

```python
img_nature_rgb = cv2.cvtColor(img_nature,cv2.COLOR_BGR2RGB)
img_nature_rgb
```

## Visualize in seaborn and matplotlib

```python
img2_grayscale
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.imshow(img2_grayscale)
plt.show()
```

```python
plt.imshow(img2_grayscale,cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(img2_grayscale,annot=True,fmt='d')
plt.show()
```

```python
img_3ri_grayscale = cv2.imread(path_3ri,cv2.IMREAD_GRAYSCALE)
img_3ri_grayscale
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(img_3ri_grayscale,annot=True,fmt='d')
plt.show()
```
