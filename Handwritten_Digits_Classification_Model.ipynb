## Model Building

```python
from keras.models import Sequential
from keras.layers import Input,Conv2D,MaxPooling2D,Dropout,BatchNormalization,Dense,Flatten
from keras.callbacks import EarlyStopping
```

```python
ytrain_scl.shape
```

```python
# Intialize the model
model = Sequential()

# Add the Input layer
model.add(Input(shape=(28,28,1),))

# Add the convolution layer
model.add(Conv2D(32,kernel_size=(3,3),activation='relu'))
# Add the pooling layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Add another convolution layer
model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
# Add another pooling layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Add the Flatten layer
model.add(Flatten())

# Add the Hidden layers
model.add(Dense(32,activation='relu'))
# Add the Dropout layer
model.add(Dropout(rate=0.2))
model.add(BatchNormalization())

# Add the Output layer
model.add(Dense(units=10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
es = EarlyStopping(monitor='val_loss',patience=5)
nn = model.fit(xtrain_scl,ytrain_scl,validation_split=0.2,epochs=15,callbacks=[es])
```

```python
import matplotlib.pyplot as plt
plt.plot(nn.history['loss'],label='Training Loss')
plt.plot(nn.history['val_loss'],label='Validation Loss')
plt.title("Learning Curve")
plt.xlabel("Number of Epochs")
plt.ylabel("Categorical_crossentropy Loss")
plt.legend()
plt.show()
```

## Out of sample predictions

```python
import cv2
def predict_images(model,img_path):
  # Read the image
  img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
  # Convert the background of image from white to black. Model was trained on black background images
  img = 255-img
  # Scaling the images
  img_scl = img.astype('float32')/255
  # Reshaping images
  img_scl = img_scl.reshape(1,28,28,1)
  # model predictions
  preds = model.predict(img_scl,verbose=1)
  # convert the probabilities into proper predictions
  final_preds = np.argmax(preds,axis=1)
  return final_preds
```

```python
paths = ['/content/Example_3.jpg','/content/Example_7.jpg','/content/Example_9.jpg']
for path in paths:
  op_preds = predict_images(model,img_path=path)
  print(f"Output predicted is: {op_preds}")
  print("================================")
```
